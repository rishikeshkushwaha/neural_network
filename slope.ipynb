```python
def cost(b):
    return (b - 4) ** 2
```

```python
cost(9)
```

```python
cost(2)
```

```python
def num_slope(b):
    h = 0.0001
    return (cost(b + h) - cost(b)) / h
```

```python
num_slope(2)
```

```python
num_slope(9)
```

```python
def slope(b):
    return 2 * (b - 4)
```

```python
slope(9)
```

```python
slope(2)
```

```python
b = 20
for i in range(100):
    b = b - 0.1 * slope(b)
    print(b)
```

```python
b = -120
for i in range(100):
    b = b - 0.1 * slope(b)
    print(b)
```
